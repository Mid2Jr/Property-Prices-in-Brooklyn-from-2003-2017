# Abstract

In this project, we analyzed the association between property sale price and various other characteristics of the property in Brooklyn between 2003 to 2017, a period of rapid gentrification. We compared the results in each of three years (2003, 2010, 2017) from different points in our time period to see how the associations between the variables and sales price change at different stages in the process of gentrification. We found that over this period there were changes in the neighborhoods and building types which were associated with higher average sale prices, but the association between sale price and other factors such as the gross square footage or the building's age remained mostly the same.

# Research Question

How have predictors of property prices in Brooklyn changed from the early 2000s (before the borough was thoroughly gentrified) to the late 2010s, after it was gentrified? In other words, are different factors considered desirable after gentrification?   

Note: we'll use the years between 2003 and 2017, as this is what we've found the best data for.

## Background and Prior Work


According to Forbes<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1), New York City is one of the most expensive cities in the world to live in, and as countless people look to move to the city each year in search of work, this affordability crisis shows no sign of improving. According to a 2021 report by the nonprofit Community Service Society, over 50% of New Yorkers are rent burdened, meaning that they spend more than 30% of their income on rent, and nearly 34% of New Yorkers spend over half their income on rent<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). While in the past, low-income residents could find housing in some less expensive areas of the city, this is often no longer the case thanks to gentrification. Gentrification describes a phenomenon in which lower-income residents are priced out of their own neighborhoods by the influx of more affluent people who are able to pay more for housing, and in turn drive up housing costs.


Gentrification has hit the borough of Brooklyn especially hard. While Brooklyn used to be seen as an undesirable, low-income neighborhood, over the past few decades it has transformed into a trendy, hip neighborhood which people are clamoring to move to <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). According to Business Insider, median rent in Brooklyn was approximately \\$1,000 cheaper than the median rent in Manhattan in 2008. By 2014, median rent was only \\$210 cheaper in Brooklyn, and during the COVID-19 pandemic the difference was as low as \\$171 <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4).

These newcomers typically differ from incumbent residents not just with respect to their income, but culturally and socially as well. According to the New York Times and the Department of City planning, “between 2000 and 2010, \[Brooklyn’s] Crown Heights and the two neighborhoods to its south and east, Flatbush and Prospect-Lefferts Gardens, all areas with large West Indian immigrant populations, each lost from 10 to 14 percent of their black populations.” <a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5). Another government report found that “The borough’s Black population declined by 8.7 percent over the last decade, while its non-Latinx white population grew by 8.4 percent.” <a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6) This trend is even more extreme within certain neighborhoods, such as Bedford-Stuyvesant. According to census data, 75% of this neighborhood was Black in 2000, but in the period from 2015-2019, only 46% was. The neighborhood’s white population rose proportionately, going from 2.4% in 2000 to nearly 30% in 2015-2019 <a name="cite_ref-7"></a>[<sup>7</sup>](#cite_note-7).
We want to investigate to what extent this social change is reflected in the characteristics of the housing market. We know that housing in Brooklyn is getting more expensive, but we want to find out how the preferences of Brooklyn’s home buyers have changed since the early 2000’s.

 

Many projects have attempted to answer this question, and we hope to build on their methods to provide insight into the housing market within New York, specifically. Algieri (2013) finds that factors like residential investment, change in population, and per capita income tend to have a positive impact on the housing price at a national level<a name="cite_ref-8"></a>[<sup>8</sup>](#cite_note-8), which can offer insight into more granular trends. While we can’t necessarily conclude that these factors will have the same impact at a smaller scale, these are questions that we want to analyze in our project.  

A Kaggle user conducted an analysis of homes in King County, WA, and found that the size of the home and the quality of the home generally have strong positive correlations with the home’s price (in line with expectations). Their analysis focused on the characteristics of the home itself (size, number of bedrooms, general quality, etc.)<a name="cite_ref-9"></a>[<sup>9</sup>](#cite_note-9), but leaves out important factors regarding the area surrounding the home, which we believe will have an influence on the price. In our analysis, we’ll try to investigate these factors in addition to analyzing variables regarding the unit itself. A Medium user conducted a similar analysis on houses in Ames, Iowa which uses methodology fairly similar to what we propose. They analyze the quality of the house and find that, unsurprisingly, higher quality homes sold for more. But, they also found a very strong relationship between the neighborhood and the sale price<a name="cite_ref-10"></a>[<sup>10</sup>](#cite_note-10). We think that we can expect a similar effect in Brooklyn, but we do think that the differences between Brooklyn and Ames, Iowa are significant enough to cause slight differences in the magnitude of this effect.   

Overall, all these projects are focused on identifying predictors of home prices at a set point in time. What we want to do is analyze how these predictors have changed over time, and discuss how these changes (if there are any) fit into the larger analysis of gentrification in the area, with the hope that our conclusions will offer insight into how housing markets can change in response to rapid demographic changes.

## Data Overview

**Dataset 1**

We're using tabular data (stored in a csv file) on sales of Brooklyn homes from 2003 to 2017. The data is available [here](https://www.kaggle.com/datasets/tianhwu/brooklynhomes2003to2017?resource=download&select=brooklyn_sales_map.csv), and it was curated by Kaggle user Tommy Wu (tianhwu). The dataset is based on sales recorded by the NYC Department of Finance, a government agency which records the sales of all properties in the city of New York. Information about the agency can be found [here](https://www.nyc.gov/site/finance/property/property-rolling-sales-data.page).  

Each row in the dataset represents the sale of a unit of real estate (sometimes a building, sometimes part of a building like a single apartment), which corresponds to a date. That is, if the same building was sold multiple times, that building will be found in one row for each time it was sold. The data dates back to 2003 and leads up to 2017, which is useful for our research question because this period from the early 2000s to late 2010s is generally recognized as the period within which Brooklyn has gentrified most significantly. To investigate the characteristics of this process and how it's reflected in the housing market, we'll use a multitude of variables.
Here's a list of some of the important variables, their datatypes, and their units:
- `sale_price` denotes the sale price of the home in US dollars. Dtype = float. This is our target/dependent variable. 
- `neighborhood` is a categorical variable indicating the neighborhood in which the building is located. Dtype = string. This variable is a proxy for the "neighborhood effects" of the surrounding area. For example, the amenities/disamenities which are available to residents. It could also be a proxy for the overall "vibe" of the property.
- `building_class_category` is a categorical variable which pertains to the property's use, in plain English. Some examples are condos, office buildings, commercial vacant land, etc. Dtype = string. Since our analysis is mainly focused on the effects on the housing market, we might eventually trim the dataset to only include buildings with residential use. However, we haven't done this yet as we recognize that the markets for both residential and nonresidential real estate are intertwined, and we aren't sure if it would be theoretically sound to exclude all nonresidential buildings.
- `year_of_sale` contains the year of the sale represented by the row. Dtype = int. This variable will obviously be used to split sales by year, but it could also be seen as a proxy for how long gentrification has been going on.
- `land_sqft` This variable represents the property's footprint, or the amount of land taken up by the property in square feet. Dtype = float
- `gross_sqft` This variable represents the total square footage of the building across all floors, or the total usable area of the property in square feet. Dtype = float
- `address` This is a categorical variable representing the property's address. The address can be proxy for a specific area of a neighborhood and can convey important information about the status of the surrounding area. For example, we can use this variable to see if a property is on a famous street such as Flatbush Avenue. Dtype = string.
- `residential_units` The number of residential units in the property. This variable (especially in conjunction with the `total_units` variable, can tell us to what extent the property is residential in nature. Dtype = int.
- `commercial_units` Similar to the previous variable, except this one representes the number of commercial units in the property. This could be a useful proxy which helps us identify the amenities available within the building, which is especially important in mixed use buildings. Dtype = int.
- `year_built` Is the year in which the unit was built. This variable clearly represents the age of a property, and can serve as a proxy for the "trendiness" of a property. Dtype = int.
- `SchoolDist` This is a categorical variable indicating which public school district the property is contained in, represented by integers (or NaNs, if the property is not in a school district). Dtype = float. Brooklyn's school districts are listed using integer codes, not English names, so there's no wrangling to be done here. 
- `HistDist` A categorical variable indicating the name of the historical district the property is found within, if any. If the property is not in any historical district, this variable takes the value "NA". Dtype = string. To wrangle the data, we may replace instances of "NA" with NaN, to simplify DataFrame operations. We may also create a new column which contains Boolean values indicating whether the property is in any historical district at all, rather than specifying which one it's in. Before we do this, we'll investigate whether certain historical districts have more expensive properties on average, or if the primary distinction is drawn simply by being in **any** historical district.


The data was already fairly clean, as all of the columns were already encoded with suitable data types, and there are very few NaN values which need to be cleaned. Any NaNs in the dataset convey valuable information, that is, all the columns we're concerned with are missing not at random (MNAR). But, we have taken a few steps to prepare the data for our analysis (such as excluding outliers), which can be found below:

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import re
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('brooklyn_sales_map.csv').drop(columns = 'Unnamed: 0')
```

```python
df.head()
```

The dataset appears to have enough observations: 390,883 over the 14 year period in question. The dataset also has 110 columns, many of which will not be particularly useful in our analysis.  

```python
df.shape
```

```python
df.columns
```

For the sake of efficiency and readability, we'll drop the columns we aren't using.

```python
df = df[['sale_price', 'neighborhood', 'building_class_category', 'year_of_sale', 'land_sqft', 'sale_date',
        'gross_sqft', 'address', 'residential_units', 'commercial_units', 'total_units', 'year_built', 'SchoolDist', 'HistDist']]
```

```python
df.head()
```

One of the columns is `year_of_sale`, which indicates which year the building was sold in. This will be what enables us to identify temporal trends in the data, so it's important that we make sure it's clean.  
Fortunately, there are no nans, and all the years seem to be encoded in integer format.

```python
df['year_of_sale'].isna().any()
```

```python
df['year_of_sale'].dtype
```

Now we'll briefly examine the distribution of the sales by year. While there are some years with more sales than others, this doesn't indicate any problem with the data, as it's an important trend which simply reflects variability in market trends. So it seems that we don't need to do any cleaning here.

```python
by_year = df.groupby('year_of_sale').count()[['sale_price']].rename({'sale_price': 'count'}, axis = 1)

by_year.plot(kind = 'bar', title='Distribution of Sale Year', ylabel='Number of Sales',
         xlabel='Year')
plt.show()
```

We can use the `year_of_sale` and the `year_built` column to create an `age` column, which will come in handy later.

```python
df['age'] = df['year_of_sale'] - df['year_built']
```

While on the subject of time, we notice that the `sale_date` column is an "object" dtype, since it contains strings. To facilitate easier operations later in our analysis, we'll change this column to contain datetime objects instead.

```python
df['sale_date'].dtype
```

```python
df['sale_date'] = pd.to_datetime(df['sale_date']) 
```

Another column which will be useful is the `building_class_category`, which describes the specific use or intended use of the building containing each property. But if we look at the unique values in it, we see there are many oddities.

```python
df['building_class_category'].unique()
```

There are some nans, and many values which are repeated due to extra whitespace. For example, '16  CONDOS - 2-10 UNIT WITH COMMERCIAL UNIT', and '16   CONDOS - 2-10 UNIT WITH COMMERCIAL UNIT'. Let's clean this up with a regex expression and drop the nans. Note that the nans here are represented by the string 'nan', so we'll simply drop them with a query.

```python
def clean_cat(cat):
    return re.sub(' +', ' ', str(cat))

df['building_class_category'] = df['building_class_category'].apply(clean_cat)
df = df[df['building_class_category'] != 'nan']
```

```python
df['building_class_category'].unique()
```

Finally, we can look at our target variable: `sale_price`. It also appears to be fairly clean, as all the values are floats representing the sale price in dollars (which is what we need), and there are no null values.

```python
df['sale_price'].isna().any()
```

```python
df['sale_price'].dtype
```

Let's take a look at the `sale_price` column to see if there are any outliers or otherwise strange values that we must account for.

```python
plt.hist(df['sale_price'], bins=20)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Sale Price')
plt.show()
```

We immediately notice that there are a number of homes which sold for exorbitant amounts of money (hundreds of millions of dollars) which cause the histogram to take this extremely skewed shape.  
In fact, the max sale price in this dataset is almost half a billion dollars.

```python
df['sale_price'].max()
```

We can see that the number of homes which sold for over 50 million is 119, and 31 homes sold for over 100 million dollars. Out of the over 300,000 observations in our dataframe, these seem to be clear outliers. So, we will exclude them from our analysis later down the line.

```python
print(f"The number of 50+ million dollar homes is {df[df['sale_price'] >= 50_000_000].shape[0]}")

print(f"The number of 100+ million dollar homes is {df[df['sale_price'] >= 100_000_000].shape[0]}")
```

For now, let's take a look at homes at or below the 90th percentile, just to get a better idea of the distribution

```python
cheaper = df[df['sale_price'] < df['sale_price'].quantile(0.90)]
```

```python
plt.hist(cheaper['sale_price'], bins=100)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Sale Price')
plt.show()
```

There are many sales which have suspiciously low sale prices, and many which are even $0. These likely reflect transfers of property (for example, a child inheriting their parents' home). As such, we'll exclude them to avoid skewing our results.  

The threshold which we're setting is \\$1000. While this is somewhat arbitrary, and probably too low, we don't want to impose a strong inductive bias on our results by exluding too many sales.

```python
df = cheaper[cheaper['sale_price'] > 1000]
plt.hist(df['sale_price'], bins=100)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Sale Price')
plt.show()
```

```python
df['sale_price'].median()
```

Now the distribution looks more normal — literally. And for reference, the median is around $445,000

### Relationships to sales price
Now that we have a better idea what each of these variables looks like, let's take a look at the relationships between sales price and some other variables of interest.

```python
time_sort = df.sort_values('sale_date')

mean_sale_price = time_sort.groupby('sale_date')['sale_price'].mean()

plt.figure(figsize=(10, 6))
plt.scatter(x = mean_sale_price.index, y = mean_sale_price.values, color='red')
plt.title('Mean Sale Price Over Time')
plt.xlabel('Sale Date')
plt.ylabel('Mean Sale Price')

plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

There seems to be an overall upward trend in sale price over time. 

This isn't too surprising. Now to investigate our research question further, we'll see if there are any changes in which characteristics are positively correlated with the sale price. To do this, we'll examine the characteristics of properties sold at the beginning of the period (2003), in the middle (2010), and at the end (2017).

```python
df_2003 = df[df['year_of_sale'] == 2003]
df_2010 = df[df['year_of_sale'] == 2010]
df_2017 = df[df['year_of_sale'] == 2017]


years = [2003, 2010, 2017]
df_filtered = df[df['year_of_sale'].isin(years)]
```

Across all years, there seems to be a weak positive correlation between sale price and square footage (both land_sqft and gross_sqft), but these are farily weak correlations, never rising above 0.34. This begs the question, how has the **housing market**, in particular, evolved? For the sake of our analysis, which is mainly concerned with other real estate markets insofar as they affect the housing market, we'll combine non-residential uses into one single label.

```python
def replace_cat(value):
    
    residential = {'02 TWO FAMILY HOMES', '03 THREE FAMILY HOMES',
       '13 CONDOS - ELEVATOR APARTMENTS',
       '12 CONDOS - WALKUP APARTMENTS',
       '15 CONDOS - 2-10 UNIT RESIDENTIAL', '04 TAX CLASS 1 CONDOS',
       '01 ONE FAMILY HOMES', '07  RENTALS - WALKUP APARTMENTS',
       '10  COOPS - ELEVATOR APARTMENTS', '14  RENTALS - 4-10 UNIT',
       '10 COOPS - ELEVATOR APARTMENTS', '09 COOPS - WALKUP APARTMENTS',
       '14 RENTALS - 4-10 UNIT', '17 CONDO COOPS', '15 CONDOS - 2-10 UNIT RESIDENTIAL',
       '09 COOPS - WALKUP APARTMENTS', '07 RENTALS - WALKUP APARTMENTS', '08 RENTALS - ELEVATOR APARTMENTS',
    '12 CONDOS - WALKUP APARTMENTS', '43 CONDO OFFICE BUILDINGS', '28 COMMERCIAL CONDOS', 
    '16 CONDOS - 2-10 UNIT WITH COMMERCIAL UNIT', '11A CONDO-RENTALS'}
    
    if value in residential:
        return value
    else:
        return 'non-residential'
```

```python
mean_sale_price = (df_filtered.groupby(['refined_category', 
                                        df_filtered['year_of_sale']])['sale_price'].mean())

mean_sale_price = mean_sale_price.unstack()
mean_sale_price.sort_values(by = 2017, ascending = False)

plt.figure(figsize=(20, 10))
mean_sale_price.plot(kind='bar', color=['blue', 'green', 'red'], width=0.8)
plt.title('Mean Sale Price by Building Class Category and Year')
plt.xlabel('Building Class Category')
plt.ylabel('Mean Sale Price')
plt.xticks(rotation=45, fontsize=5)

plt.legend(title='Year', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

While almost all property types have gotten more expensive, there is one exception. Notably, 4-10 unit rentals have gotten cheaper from 2010 to 2017, even dropping below 2003 levels. # Ethics & Privacy

The primary concern when it comes to privacy is simply that this dataset makes it easy to identify information regarding a certain home, and therefore about the people who live in that home. Since this data is all publically available, we aren't publishing any private information by using the dataset. However, we have made sure to avoid using any analysis of individual properties (instead aggregating across building types, neighborhoods, etc.) so as to maintain as much anonymity for the residents as possible. In addition to this privacy concern, there are plenty of potential biases which may arise from our dataset. One potential bias is is that some sales may be absent from the dataset due to the way that the New York government collects the data. It only includes properties classified in New York's tax classes 1, 2, and 4, excluding class 3. Class 3 "Includes property with equipment owned by a gas, telephone or electric company," according to the New York Department of Finance. This shouldn't cause a huge problem for our project, but we should still make note of it, since it's possible that some buildings contain this equipment in addition to housing units which we otherwise want to analyze. Another potential bias is that any sales done illegally, or "under the table," won't be present in our dataset. There isn't much we can do to account for this bias, but we also expect that it won't pose much of an issue since such deals aren't particularly common. The last source of bias we'd like to note is that with this research question, we're inherently working with a "chicken or the egg" problem, in the sense that it's difficult to say whether changes in the characteristics of Brooklyn's homes cause gentrification, or if gentrification causes the changes. We've tried to get around this bias by framing the question in a way that simply explores associations between the price and other factors regarding the properties, but we recognize that this prevents us from attempting to establish causal relationships between either variable. Finally, we are wary of using this project to propose any policy changes, as this may get in the way of our project having an equitable impact. Since we haven't established causality, we don't want to conclude something like "building more schools causes prices to rise, so we shouldn't build schools in poor communities." We've tried to address this issue in our analysis by discussing associations between variables, with the intention of identifying areas for more rigorous economic research.
